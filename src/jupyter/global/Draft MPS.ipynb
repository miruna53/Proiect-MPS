```python
import pandas as pd
import numpy as np
import os
import glob
import csv
```

```python
path = os.getcwd()
path += '\mps-global\MPS-GLOBAL'
print(path)

csv_files = glob.glob(os.path.join(path, "*.csv"))

print(type(csv_files))
print(len(csv_files), '\n')

def convert(string):
    li = string.split(",")
    
    while("" in li):
        li.remove("")
        
    return li

f_measure_error = []
# loop over the list of csv files
for file in csv_files: #[0:5]
    optimal_threshold = 0
    thresholds = []
    f_measures = []
      
    with open(file, "r") as f:
        reader = csv.reader(f, delimiter="\t")
        for i, line in enumerate(reader):
            if i == 0:
                thresholds = [float(num) for num in convert(line[0])]
                optimal_threshold = thresholds[0]
                thresholds = thresholds[1:]
#                 print("Thresholds: ", thresholds, '\n') 
            if i == 1:
                f_measures = [float(num) for num in convert(line[0])]
#                 print("F-measures: ", f_measures, '\n')
                
    print("Optimal threshold: ", optimal_threshold)
    print("Mean: ", np.mean(thresholds))
    
    optimal_f_measure = f_measures[int(np.floor(optimal_threshold * 255))]
    our_f_measure = f_measures[int(np.floor(np.mean(thresholds) * 255))]
    print("Optimal f-measure: ", optimal_f_measure)
    print("Our f-measure: ", our_f_measure, '\n')
    
    f_measure_error.append(optimal_f_measure - our_f_measure)

print("Mean of errors: ", np.mean(f_measure_error))
```
